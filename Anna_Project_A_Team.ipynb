```python
major_list = ['computer science','data science','statistics','quantitative',
            'mathematics','economics','liberal arts','science','business',
            'engineer','engineering','psychology','biology','medicine',
            'marketing','business analysis','chemical engineering','chemistry',
            'public health','public relations']
len(major_list)
```

```python
for m in major_list:
    counter = data["description"].str.lower().str.contains(m).value_counts()
    print(counter)
```

```python
data["description"].str
```

```python
frequency_false=[]
frequency_true=[]

for m in major_list:
    counter = data["description"].str.lower().str.contains(m).value_counts()
    #print(m, ":",counter)
    frequency_false.append(counter[False])

for f in frequency_false:
    frequency_true.append(abs(f - 6953))
```

```python
frequency_true
```

```python
major_df=pd.DataFrame()
major_df["major"]  = major_list
major_df["major_freq"]  = frequency_true
```

```python
lower_position = data['position'].str.lower()
category = []
for position in lower_position:
    if "data scientist" in position or 'data science' in position or "machine learning" in position:
        category.append('data scientist')
    elif "analyst" in position or "analytics" in position:
        category.append('analyst')
    elif "engineer" in position or "artificial intelligence" in position or "ai" in position:
        category.append('engineer')
    else:
        category.append('other')
```
